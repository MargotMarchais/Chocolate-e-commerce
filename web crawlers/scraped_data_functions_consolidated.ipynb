<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">Transform variables: product weight</h3>
</div>

Focus : Lindt, Abtey and Maison du Chocolat datasets have no values in the product weight column since the information is not directly accessible on their website. However, the information may be available in the product name (e.g."Sachet LINDOR 500g"). Hence, we create a little code to extract that information from the product name

```python
# Overview of the product names: 
    # We cannot however extract all numbers from the product name, because sometimes
    # it refers to the % of chocolate in the product instead of the product weight (eg. Moulage LAPIN OR Noir 70% Cacao 100g) 
liste = [df_abtey, df_lindt, df_maison_du_chocolat, df_castelain]

for df in liste:
    
    # The weight is often mentioned at the end of the product name: We will extract the last 6 characters of the product name
    # and extract the numbers ('weight' column) and the letters ('unit')
    df["weight"] = df["product_name"].str[-6:].str.replace(r'\D+', '', regex=True)
    df["unit"] =  df["product_name"].str[-6:].str.replace(r'\d+', '', regex=True)
    
    # Cleaning the results : sometimes, the product name does not contain weight information (eg: Sac en papier brun & doré Lindt - Petit Modèle)
    # So we will clean the results
    df["unit"] = np.where(df["unit"].str.contains('kg'), 'kg',
                            np.where(df["unit"].str.contains('KG'), 'kg',
                                np.where(df["unit"].str.contains('g'), 'g', None)
                                    )
                           )
    df["weight"] = df["weight"].apply(pd.to_numeric)
    df["weight_grams"] = np.where(df["unit"]=="kg", df["weight"]*1000, df["weight"])
    
df_abtey.tail(3) 
```

```python
# Remove useless variables
df_lindt = df_lindt.loc[:, ['timestamp_photo', 'product_name', 'image_url', 'product_url', 'price', 'date_photo', 'weight_grams']]

df_maison_du_chocolat = df_maison_du_chocolat.loc[:, ['timestamp_photo', 'product_name', 'image_url', 'product_url',
       'quantity', 'price', 'rating', 'product_description', "weight_grams", 'date_photo']]

df_castelain = df_castelain.loc[:, ['id', 'product_url', 'on_sale', 'quantity',
       'low_stock_alert', 'price_text', 'available_for_order', 'date_update',
       'link_rewrite', 'meta_description', 'meta_title', 'product_name',
       'category', 'category_name', 'price_no_reduction', 'has_discount',
       'image_url', 'brand', 'format', 'saveur', 'teneur_cocoa',
       'timestamp_photo', 'date_photo', 'weight_grams']]

df_abtey = df_abtey.loc[:, ['id_product', 'product_name', 'product_url', 'image_url',
       'product_price_text', 'has_discount', 'discount_percentage',
       'discount_amount_text', 'product_price', 'regular_price',
       'regular_price_text', 'timestamp_photo', 'date_photo','weight_grams']]
```

```python
# Convert all quantities to grams
for df in list2_df:
    df["weight_grams"] = np.where(df["unit"]=="kg",
                                 df["weight"]*1000,
                                 df["weight"])
df_jeff_de_bruges.head()
```
