Focus : Lindt dataset has no values in the product weight column since the information is not directly accessible on their website. However, the information may be available in the product name (e.g."Sachet LINDOR 500g"). Hence, we create a little code to extract that information from the product name

```python
# Overview of the Lindt product names: 
    # We cannot however extract all numbers from the product name, because sometimes
    # it refers to the % of chocolate in the product instead of the product weight (eg. Moulage LAPIN OR Noir 70% Cacao 100g) 
print(df_lindt[["product_name", "product_weight"]].tail(8))

# The weight is often mentioned at the end of the product name: We will extract the last 6 characters of the product name
    # and extract the numbers ('weight' column) and the letters ('unit')
df_lindt["weight"] = df_lindt["product_name"].str[-6:].str.replace(r'\D+', '', regex=True)
df_lindt["unit"] = df_lindt["product_name"].str[-6:].str.replace(r'\d+', '', regex=True)

# Cleaning the results : sometimes, the product name does not contain weight information (eg: Sac en papier brun & doré Lindt - Petit Modèle)
    # So we will clean the results
df_lindt["unit"] = np.where(df_lindt["unit"].str.contains('kg'), 'kg', 
                            np.where(df_lindt["unit"].str.contains('g'), 'g', None)
                           )
df_lindt["weight"] = df_lindt["weight"].apply(pd.to_numeric)
df_lindt["weight_grams"] = np.where(df_lindt["unit"]=="kg", df_lindt["weight"]*1000, df_lindt["weight"])
df_lindt.tail(8)
```

```python
df_lindt.columns
df_lindt = df_lindt.loc[:, ['timestamp_photo', 'product_name', 'image_url', 'product_url', 'price', 'date_photo', 'weight_grams']]
df_lindt.head()
```

```python
# Maison du Chocolat : Product weight
df_maison_du_chocolat.head()
df_maison_du_chocolat["weight"] = df_maison_du_chocolat["product_name"].str[-6:].replace(r'\D+', '', regex = True)
df_maison_du_chocolat["unit"] = df_maison_du_chocolat["product_name"].str[-6:].replace(r'\d+', '', regex = True)
df_maison_du_chocolat["unit"] = np.where(df_maison_du_chocolat["unit"].str.contains('kg'), 'kg', 
                            np.where(df_maison_du_chocolat["unit"].str.contains('g'), 'g', None)
                           )
df_maison_du_chocolat["weight"] = df_maison_du_chocolat["weight"].apply(pd.to_numeric)
df_maison_du_chocolat["weight_grams"] = np.where(df_maison_du_chocolat["unit"]=="kg", 
                                                 df_maison_du_chocolat["weight"]*1000, 
                                                 df_maison_du_chocolat["weight"])
```

```python
# Remove useless variables
df_maison_du_chocolat.columns
df_maison_du_chocolat = df_maison_du_chocolat.loc[:, ['timestamp_photo', 'product_name', 'image_url', 'product_url',
       'quantity', 'price', 'rating', 'product_description', "weight_grams", 'date_photo']]
df_maison_du_chocolat.tail()
```

```python
# Obter datasets : Obtain a numeric product weight (unit standardized to grams)
list2_df = [df_jeff_de_bruges, df_alain_ducasse, df_franprix]

for df in list2_df:
    df["weight"] = df["product_weight"].str.replace(r'\D+', '', regex=True)
    df["unit"] = df["product_weight"].str.replace('\d+', '')
    df["weight"] = df["weight"].apply(pd.to_numeric)
    
df_jeff_de_bruges.head()
```

```python
# Convert all quantities to grams
for df in list2_df:
    df["weight_grams"] = np.where(df["unit"]=="kg",
                                 df["weight"]*1000,
                                 df["weight"])
df_jeff_de_bruges.head()
```

```python
# Remove useless columns
for df in list2_df:
    df = df.drop(columns = ["product_weight", "weight", "unit"], inplace =True)

df_jeff_de_bruges.head()
```

In the Alain Ducasse and Maison du Chocolat datasets, the quantity information may be available but has to be reformatted. 

```python
# Alain Ducasse dataset: get rid of the string characters in the "quantity" variable (eg: "27 pièces" should turn into "27")
df_alain_ducasse["quantity"] = df_alain_ducasse["quantity"].str.replace(r'\D+', '', regex = True)
df_alain_ducasse["quantity"] = df_alain_ducasse["quantity"].apply(pd.to_numeric)
df_alain_ducasse.head()
```

```python
# Maison du chocolat : Quantities
df_maison_du_chocolat["quantity"] = df_maison_du_chocolat["product_name"].replace(r'\D+', '', regex = True)

df_maison_du_chocolat["quantity"] = np.where(df_maison_du_chocolat["quantity"] == df_maison_du_chocolat["weight_grams"],
                                            None,
                                            df_maison_du_chocolat["quantity"])

df_maison_du_chocolat["quantity"] = df_maison_du_chocolat["quantity"].apply(pd.to_numeric)
df_maison_du_chocolat.tail()
```

```python
# Price variable : transform the "," into "."
df_jeff_de_bruges["price"] = df_jeff_de_bruges["price"].str.replace(",", ".")
df_lindt["price"] = df_lindt["price"].str.replace(",", ".")
df_alain_ducasse["price"] = df_alain_ducasse["price"].str.replace(",", ".")
df_maison_du_chocolat["price"] = df_maison_du_chocolat["price"].str.replace(",", ".")

df_jeff_de_bruges.head()
```

```python
# Price variable: remove "à partir de xxx"
df_alain_ducasse["price"] = df_alain_ducasse["price"].str.replace('à partir de ', '')
```

```python
# Price variable: Remove the currency € sign at the end of the string
no_currency = lambda x: float(x[:-2])

df_jeff_de_bruges["price"] = df_jeff_de_bruges["price"].apply(no_currency)
df_lindt["price"] = df_lindt["price"].apply(no_currency)
df_alain_ducasse["price"] = df_alain_ducasse["price"].apply(no_currency)
df_maison_du_chocolat["price"] = df_maison_du_chocolat["price"].apply(no_currency)
```
